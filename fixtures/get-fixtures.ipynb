```python
import pandas as pd
from datetime import datetime
import json
from collections import OrderedDict
import numpy as np
```

```python
df0_epl = pd.read_csv("E0.csv")
df1_epl = pd.read_csv("E1.csv")
df_epl = df0_epl.append(df1_epl)

df0_bdl = pd.read_csv("D0.csv")
df1_bdl = pd.read_csv("D1.csv")
df_bdl = df0_bdl.append(df1_bdl)

df0_fra = pd.read_csv("F0.csv")
df1_fra = pd.read_csv("F1.csv")
df_fra = df0_fra.append(df1_fra)

df0_ita = pd.read_csv("I0.csv")
df1_ita = pd.read_csv("I1.csv")
df_ita = df0_ita.append(df1_ita)

df0_spa = pd.read_csv("SP0.csv")
df1_spa = pd.read_csv("SP1.csv")
df_spa = df0_spa.append(df1_spa)
```

```python
df_fra.dropna(how='all', inplace=True)
df_ita.dropna(how='all', inplace=True)
df_spa.dropna(how='all', inplace=True)
```

```python
df_epl['Date'] = df_epl['Date'].apply(lambda x : datetime.strptime(x, '%d/%m/%y'))
df_bdl['Date'] = df_bdl['Date'].apply(lambda x : datetime.strptime(x, '%d/%m/%y'))
df_fra['Date'] = df_fra['Date'].apply(lambda x : datetime.strptime(x, '%d/%m/%y'))
df_ita['Date'] = df_ita['Date'].apply(lambda x : datetime.strptime(x, '%d/%m/%y'))
df_spa['Date'] = df_spa['Date'].apply(lambda x : datetime.strptime(x, '%d/%m/%y'))
```

```python
mufc = df_epl[df_epl['HomeTeam'].str.contains("Man United")].append(df_epl[df_epl['AwayTeam'].str.contains("Man United")])
liverpool = df_epl[df_epl['HomeTeam'].str.contains("Liverpool")].append(df_epl[df_epl['AwayTeam'].str.contains("Liverpool")])
mcfc = df_epl[df_epl['HomeTeam'].str.contains("Man City")].append(df_epl[df_epl['AwayTeam'].str.contains("Man City")])
arsenal = df_epl[df_epl['HomeTeam'].str.contains("Arsenal")].append(df_epl[df_epl['AwayTeam'].str.contains("Arsenal")])
spurs = df_epl[df_epl['HomeTeam'].str.contains("Tottenham")].append(df_epl[df_epl['AwayTeam'].str.contains("Tottenham")])
chelsea = df_epl[df_epl['HomeTeam'].str.contains("Chelsea")].append(df_epl[df_epl['AwayTeam'].str.contains("Chelsea")])

bayern = df_bdl[df_bdl['HomeTeam'].str.contains("Bayern Munich")].append(df_bdl[df_bdl['AwayTeam'].str.contains("Bayern Munich")])
dortmund = df_bdl[df_bdl['HomeTeam'].str.contains("Dortmund")].append(df_bdl[df_bdl['AwayTeam'].str.contains("Dortmund")])

inter = df_ita[df_ita['HomeTeam'].str.contains("Inter")].append(df_ita[df_ita['AwayTeam'].str.contains("Inter")])
juventus = df_ita[df_ita['HomeTeam'].str.contains("Juventus")].append(df_ita[df_ita['AwayTeam'].str.contains("Juventus")])
milan = df_ita[df_ita['HomeTeam'].str.contains("Milan")].append(df_ita[df_ita['AwayTeam'].str.contains("Milan")])
roma = df_ita[df_ita['HomeTeam'].str.contains("Roma")].append(df_ita[df_ita['AwayTeam'].str.contains("Roma")])

madrid = df_spa[df_spa['HomeTeam'].str.contains("Real Madrid")].append(df_spa[df_spa['AwayTeam'].str.contains("Real Madrid")])
barcelona = df_spa[df_spa['HomeTeam'].str.contains("Barcelona")].append(df_spa[df_spa['AwayTeam'].str.contains("Barcelona")])
atletico = df_spa[df_spa['HomeTeam'].str.contains("Ath Madrid")].append(df_spa[df_spa['AwayTeam'].str.contains("Ath Madrid")])
sevilla = df_spa[df_spa['HomeTeam'].str.contains("Sevilla")].append(df_spa[df_spa['AwayTeam'].str.contains("Sevilla")])

psg = df_fra[df_fra['HomeTeam'].str.contains("Paris SG")].append(df_fra[df_fra['AwayTeam'].str.contains("Paris SG")])
lyon = df_fra[df_fra['HomeTeam'].str.contains("Lyon")].append(df_fra[df_fra['AwayTeam'].str.contains("Lyon")])
```

```python
clubs = {'mufc':mufc, 'liverpool':liverpool, 'mcfc':mcfc, 'arsenal':arsenal, 'spurs':spurs, 'chelsea':chelsea,
        'bayern':bayern, 'dortmund':dortmund, 'inter':inter, 'juventus':juventus, 'milan': milan, 'roma':roma,
        'madrid':madrid, 'barcelona':barcelona, 'atletico': atletico, 'psg': psg, 'lyon': lyon}

for club in clubs:
    club_fixtures = list()
    for idx, val in clubs[club].iterrows():
        d = OrderedDict()
        d['date'] = val[1]
        d['club1'] = val[2]
        d['club1score'] = int(val[4])
        d['club2'] = val[3]
        d['club2score'] = int(val[5])
        club_fixtures.append(d)
    
    club_fixtures = sorted(club_fixtures, key=lambda x : x['date'])
    
    for i in club_fixtures:
        i['date'] = i['date'].strftime('%m/%d/%Y')
        
    with open(club,'w') as f:
        json.dump(club_fixtures, f)
```
